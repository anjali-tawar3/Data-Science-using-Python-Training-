```python
from googlesearch import search

#title=input('what do you want to search')
title='data science'
data=search(title, num_results=10)

all_data=[]

for i in data:
    all_data.append(i+'\n')

#mode= w : only write
file=open(f'{title}.txt',mode='w')
file.writelines(all_data)
file.close()
print('done')
```

```python
from googlesearch import search

title='data science'
data=search(title, num_results=10)

all_data=[]

for i in data:
    all_data.append(i+'\n')

#mode=x : it only creates new file
file=open(f'{title}.txt',mode='x')
file.writelines(all_data)
file.close()
print('done')
```

```python
from googlesearch import search

title='chat GPT trending new'
data=search(title, num_results=2)

all_data=[]

for i in data:
    all_data.append(i+'\n')

#mode= a : it will append new line(if the file doesn't exist it will make the file)
file=open(f'{title}.txt',mode='a')
file.write('====================================================\n')
file.writelines(all_data)
file.close()
print('done')
```

```python
f = open('day_4.txt', 'w+')  
print(f.write('THIS IS A LINE'))
f.seek(0) #sending the cursor at the index of 0
print(f.read())
f.close()
print('done')
```

```python
#how to create a copy of any file
f1 = open('day_4.txt', 'r')
all_data=f1.read()

f2=open('day_4_copy.txt','w')
f2.write(all_data)
f1.close()
f2.close()
print('done')
```

```python
url='''https://picsum.photos/1080/720'''
import requests as r
n=6
for i in range(n):
    img_data=r.get(url).content
    f=open(f'image_{i}.jpg','wb')
    f.write(img_data)
    f.close()
    from IPython import display as d
    display(d.Image(f'image_{i}.jpg', width=200, height=100))
```
