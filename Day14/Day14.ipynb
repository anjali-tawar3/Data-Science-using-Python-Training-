```python
# internet required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
# sns.countplot
# sns.barplot
# sns.boxplot
# sns.scatterplot
# sns.catplot
# sns.pairplot
# sns.heatmap
# sns.regplot
# sns.lmplot
# sns.lineplot
# sns.kdeplot
# sns.violinplot
#sns.relplot
#sns.histplot
#sns.displot
```

```python
import matplotlib.pyplot as plt
sns.countplot(data= df_titanic, x='sex') # hue: further divide: bases on categorical data
plt.show()
```

```python
# sns.countplot
sns.countplot(data= df_titanic, x='sex', hue= 'alive', palette= sns.color_palette('mako',2)) # hue: further divide: bases on categorical data
plt.title('Titanic Survival Analysis(Based on Sex)')
plt.show()
```

```python
# juilius Ai 
```

```python
# sns.barplot
sns.barplot(data=df_titanic, x='class', y='fare', hue='alive',palette= sns.color_palette('mako',2)) #estimator by default show mean
plt.title('Titanic Fare Prized Analysis(Based on Class)')
```

```python
# confidence interval (CI) : by default 95%
sns.barplot(data=df_titanic, x='class', y='fare', hue='alive',palette= sns.color_palette('mako',2),errorbar=('ci',0)) # errorbar refers to the CI value
plt.grid(color= 'r',alpha= 0.1)
plt.title('Titanic Fare Prized Analysis(Based on Class)')
```

```python
# MOST IMPORTANT PLOT IN SEABORN
# sns.boxplot : to detect outliers and check data distribution
# to detect outliers:
# range : min to max(19-22) :35,28,45(outliers : fliers)
# Q2: Quantile: 50%
# Q1: Quantile: 25%
# Q3: Quantile: 75%
# IQR: Q3-Q1
# emperical rule:
# UW(upper whiskers)= Q3 + 1.5 * IQR
# LW(lower whiskers)= Q1 - 1.5 * IQR
```

```python
sns.boxplot(data=df_titanic,x='class', y='fare', hue='sex',palette= sns.color_palette('mako',2))
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(data=df_titanic,y='class', x='fare', hue='sex',palette= sns.color_palette('mako',2),orient='h')
```

```python
# sns.violinplot
sns.violinplot(data= df_titanic,x='class',y='fare', hue='sex')
```
