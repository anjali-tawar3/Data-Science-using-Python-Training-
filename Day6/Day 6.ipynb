```python
class ATM:
    def __init__(self):
        print('Object Created Successfully')
        self.pin= 1010 
        self.balance= 5000 


    def change_pin(self):
        self.pin=input(int('enter new pin'))
        print('your PIN has changed!!!')

    def show_balance(self):
        print(f'''Your current Balance is Rs {self.balance}''')

    def atm_performance_status(self):
        print('ATM working fine 100%')
```

```python
icici=ATM()
```

```python
icici.atm_performance_status
```

```python
class ATM_version1:
    def __init__(self):
        print('Object Created Successfully')
        self._pin= 1010 #protected variable :using underscore(_) will make the instance protected and be hidden from the user
        self._balance= 5000 #protected variable


    def change_pin(self): #public method
        self._pin=input(int('enter new pin'))
        print('your PIN has changed!!!')

        

    def show_balance(self):
        print(f'''Your current Balance is Rs {self._balance}''')

        

    def atm_performance_status(self):
        print('ATM working fine 100%')
```

```python
axis=ATM_version1()
```

```python
axis._balance
```

```python
class ATM_version2:
    def __init__(self):
        print('Object Created Successfully')
        self.__pin= 1010 #private variable :using double underscore(_ _) will make the instance private and be hidden from the user
        self.__balance= 5000 #private variable


    def change_pin(self): #public method
        self.__pin=input(int('enter new pin'))
        print('your PIN has changed!!!')

        

    def show_balance(self):
        print(f'''Your current Balance is Rs {self.__balance}''')

        

    def __atm_performance_status(self):
        print('ATM working fine 100%')
```

```python
hdfc=ATM_version2()
```

```python
hdfc._ATM_version2__balance
```
