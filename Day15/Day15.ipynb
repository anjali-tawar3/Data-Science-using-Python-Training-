```python
plt.scatter(df['Study_hrs'],df['Marks_obtained'],marker='x')
plt.xlabel('Study Hrs')
plt.ylabel('Marks Obtained')
plt.title('Study vs Marks')
```

```python
#import of module
from sklearn.linear_model import LinearRegression

#object model creation 
model=LinearRegression()

#X,y
X=df[['Study_hrs']]
y=df['Marks_obtained']
#training of the model
model.fit(X,y)
```

```python
x = [[5.5]]
w = model.coef_
c = model.intercept_

print(w,c)
```

```python
# y = m*x+c
std_hrs = [[4.5]]
marks = w*std_hrs + c
print('Predicted Marks:', marks[0][0]) # [][] are used to remove the dataframe
```

```python
import warnings
warnings.filterwarnings('ignore')
std_hrs = input('enter study hours:')
std_hrs = [[float(std_hrs)]]
# if std_hrs >= 24:
#     print('not valid')
# else:
#     std_hrs = std_hrs
    
ans = model.predict(std_hrs)
marks= round(ans[0],2)

if marks > 100:
    marks= 100
elif marks < 0:
    marks= 0
else:
    marks = marks
print(f'Your predicted marks is:,{marks}')
```
